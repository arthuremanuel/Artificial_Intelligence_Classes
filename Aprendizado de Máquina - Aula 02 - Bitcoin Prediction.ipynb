**Correlation measures the relationship between two variables.**
* Values Between -1 and 1
 * 1 = there is a perfect linear relationship between the variables 
 * 0 = there is no linear relationship between the variables
 * -1 = there is a perfect negative linear relationship between the variables

**There is an important difference between correlation and causality:**
* Correlation is a number that measures how closely the data are related
* Causality is the conclusion that x causes y.

**Correlation Does Not Imply Causality**

* During the summer, the sale of ice cream at a beach increases
* Simultaneously, drowning accidents also increase as well
* Does this mean that increase of ice cream sale is a direct cause of increased drowning accidents?

# Decisão de Projeto:
* Desenvolver um modelo de ML para previsão do valor de fechamento (Close) do Bitcoin no dia seguinte.
* Este (Close) será o nosso "alvo" a ser previsto [Aprendizado Supervisionado - Regressão]
  * Os dados serão dividos, portanto em: (1) dados de entrada, sem a coluna 'Close' e (2) dados de saída [rótulos], com a coluna 'Close', que serão o resultado esperado do algoritmo

**O que iremos prever?**
O preço de fechamento do BTC no dia seguinte

```python
df['Close_Tomorrow'] = df['Close'].shift(-1)
df.head()
```

```python
#Data Cleaning
#Most Machine Learning algorithms cannot work with missing features.
#We have three options:
#Get rid of the corresponding lines
#Get rid of the whole attribute.
#Set the values to some value (zero, the mean, the median, etc.) - Training set data ONLY

df = df.dropna() ## remove as linhas com NaN ##
df.head()
```

**Divisão entre conjunto de treinamento e teste** <br>
Esta separação é importante para podermos testar o modelo produzido com dados dos quais ele não conhece.
Isso também pode ser feito com uma biblioteca: <br>
from sklearn.model_selection import train_test_split 

```python
testStartDate = '01-06-2022'

train  =  df[:testStartDate]
test    = df[testStartDate:]
```

```python
train
```

```python
test
```

**Feature Scaling** <br>
Machine Learning algorithms don’t perform well when the input numerical attributes have
very different scales. Ex: Stock Prices and Volume <br>
There are two common ways to get all attributes to have the same scale: min-max scaling [values are shifted and rescaled so that they end up ranging from 0 to 1] and standardization [subtracts the mean value, and then it divides by the variance so that the resulting distribution has unit variance]. Unlike
min-max scaling, standardization does not bound values to a specific range, which may be a problem for
some algorithms (e.g., neural networks often expect an input value ranging from 0 to 1).

```python
# APENAS DADOS DE TREINAMENTO #            
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaled_features_train = scaler.fit_transform(train)
scaled_features_train
```

```python
#recriando o DF de treinamento
df_scaled_features_train = pd.DataFrame(scaled_features_train, index=train.index, columns=train.columns)
df_scaled_features_train
```

**Treinando o modelo**


```python
#Separando features de labels
X_train = df_scaled_features_train.drop(["Close_Tomorrow"], axis=1) 
y_train = df_scaled_features_train['Close_Tomorrow']

####################################
####################################
####################################
# Considerar cross validation futuramente -- Increasing Window Cros Validation
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.TimeSeriesSplit.html
#https://scikit-learn.org/stable/modules/cross_validation.html#time-series-split
####################################
####################################
####################################

from sklearn.linear_model import LinearRegression
model = LinearRegression()
linreg = model.fit(X_train , y_train)

##Outros modelos
##Redes Neurais
##from sklearn.neural_network import MLPRegressor
##model = MLPRegressor(hidden_layer_sizes=(50,50,50), max_iter=500, early_stopping = True)
##mlp = model.fit(X_train , y_train)

##SVM
##
##
```

```python
#############################################################################
##################    NORMALIZAR OS DADOS DE TESTE E PREVER #################
#############################################################################
scaled_features_test = scaler.transform(test)
df_scaled_features_test = pd.DataFrame(scaled_features_test, index=test.index, columns=test.columns)

X_test = df_scaled_features_test.drop(["Close_Tomorrow"], axis=1) 
y_test = df_scaled_features_test['Close_Tomorrow']

y_pred = model.predict(X_test)
y_pred
```

```python
df_scaled_features_test['Close_Tomorrow'] = y_pred  # Adicionando o que foi previsto ao DF de teste
# Desnormalizando os dados para a escala original
y_pred_unorm = scaler.inverse_transform(df_scaled_features_test)

test_pred = pd.DataFrame(y_pred_unorm, index=test.index, columns=test.columns)

test_pred
```

```python
test['Close_Pred'] = test_pred['Close_Tomorrow']
test[['Close_Pred','Close_Tomorrow']]
```

# Cuidados!
* Previsão acertada no passado não garante previsão correta para o futuro! - Problema do peru no dia de ação de graças (indução de David Hume ou Cisne Negro).
* Alguns modelos se adaptam demais aos dados ou necessitam de dados normalizados.
* É necessário trabalhar melhor com  os conjuntos de dados e utilizar mais etapas de desenvolvimento para que o modelo seja confiável.
  * Melhorar a validação contra riscos, uso de conjunto de dados de validação e usar Cross Validation, além de trabalhar mais com a extração de características dos dados (indicadores técnicos).
  * Os dados possuem escalas diferentes, o que pode impactar alguns algoritmos de ML. Para resolver isso, é necessário uma etapa prévia de normalização dos dados, que será discutida posteriormente. 
* Pode-se utilizar modelos mais interessantes e eficientes para não-linearidades:
  * SVM, Redes Neurais, etc.
